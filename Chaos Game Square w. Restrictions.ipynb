```python
#Start by defining initial vertices and graphing
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random

distance = .5
vertex_x = [0,0,1,1]
vertex_y = [0,1,1,0]
vertex_comb = [vertex_x, vertex_y]

# now create random first point wihin triangle
# Use random weights of each vertex to get WA random point. 
# NOTE: IM NOT SURE THIS ACTUALLY REPRESENTS A UDF RANDOM SELECTION. 

random_seed=[]
for i in range(4):
    random_seed.extend([random.random()])
x = [a*b for a,b in zip(vertex_x,random_seed)]
initial_x = sum(x)/sum(random_seed)

y = [a*b for a,b in zip(vertex_y,random_seed)]
initial_y = sum(y)/sum(random_seed)

vertex_x.extend([initial_x])
vertex_y.extend([initial_y])

#Now select random vertex to 'approach', then create point at a weighted midpoint to it. 
last_rand = -2
for i in range (100000):
    if last_rand == 0:
        restrict = last_rand+3 
    else: 
        restrict = last_rand-1
    while True: 
        rand_vertex = random.randint(0,3)
        if rand_vertex != restrict:
            break
    vertex_x.extend([(vertex_x[rand_vertex]*distance+vertex_x[len(vertex_x)-1]*(1-distance))])
    vertex_y.extend([(vertex_y[rand_vertex]*distance+vertex_y[len(vertex_y)-1]*(1-distance))])
    last_rand = rand_vertex

plt.rcParams['figure.figsize'] = 10, 10
plt.scatter(vertex_x,vertex_y, s=.25)
plt.show()
```
